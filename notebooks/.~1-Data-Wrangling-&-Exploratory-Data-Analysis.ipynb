## Introduction<a id='1.2_Introduction'></a>

***Lights................!!! Camera....................!!!Action!***

A widely known form of communication and highly influential art, film occured merely by accident, persistence, and experimentation. The inception of storytelling dates as far back as to ancestors' carvings found on the walls located in caves to monuments built during ancient times to Shakespeare's masterpieces, and now the world is able to produce 8k quality. Conceived by Max Wetheimer, the phi phenomenon states that if frames were displayed sequentially at a fifth of a second then the brain won't be able to register the black space in between pictures similarly to a flip book. Motion picture orginated from this notion. The first form of this optical illusion of moving photographs came from using a zoetrope.    

Cinema didn't occur from an individual striving towards his/her plan with a clear vision, but from a series of cheerful, unexpected events and technological advancements. During the development of photography, Joseph Nicephore Niepce used a plate, camera obscura (dark box with lens that projects the image onto a screen), and light exposure to develop the first ever photograph. From there to Thomas Edison's Kinetograph/Kinetoscope to Lumiere Brothers' cinematographe to Georges Melies editing, film has arrived.  

With motion picture camera booming in the late 1800s, films started getting produced at a rapid rate as it attracted lots of attention. As technology progressed, Edison and others took all efforts to expand this field.    

Film wasn't created, it was stumbled upon through trial and error. With this discovery, extraordinary movies like Shutter Island, Inception, The Dark Knight, Memento, and Harry Potter were watched.

### Columns/Features<a id='1.5.1_Columns_/_Features'></a>
- **adult**: specifies whether the film is X-Rated or Adult
- **belongs_to_collection**: identifies the movie series/franchise a film belongs to (Toy Story, Harry Potter)
- **budget**: budget of the film in dollars
- **genres**: genres/classes associated with each movie
- **homepage**: official homepage of the movie
- **id**: movie ID
- **imdb_id**: IMDB ID of the film 
- **original_language**: the language the film was officially shot in
- **original_title**: the name of the movie upon release
- **overview**: brief summary of the plot
- **popularity**: popularity score that indicates how well known the film is
- **poster_path**: URL of the movie poster image
- **production_companies**: companies involved in the making of the film
- **production_countries**: countries where the movie was released/shot in 
- **release_date**: theatrical release date 
- **revenue**: total return of the film
- **runtime**: number in minutes length of film 
- **spoken_languages**: languages spoken in the film 
- **status**: state of film (Released, Rumored, Post-Production, etc.)
- **tagline**: punch line/catch phrase
- **video**: expresses whether or not if there is a video of the film present on TMDB
- **vote_average**: average score of the movie 
- **vote_count**: the number of votes by users

```python
def clean_col(df,col):
    """ 
    @ params: dataframe, column
    removes irrelevant information and outputs only list of the names in each dictionary
    """
    df[col] = df[col].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else[])
```

```python
cols = ['genres', 'production_countries', 'spoken_languages', 'belongs_to_collection']
for col in cols:
    clean_col(movies, col)
```

```python
def scatterplots(columns, ncol=None, figsize=(15, 8)):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.5, hspace=0.6)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        ax.scatter(x=col,  y= 'vote_average', data=movies, alpha=0.5)
        ax.set(xlabel=col, ylabel='Vote Average')
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```
